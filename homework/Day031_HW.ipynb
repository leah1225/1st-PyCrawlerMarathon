### 尋找一個網站是有作 Header 檢查的（Hint: 有加跟沒有加回傳結果不一樣）

```python
import requests
r = requests.get("https://www.zhihu.com/topic/19550228/hot")
response = r.text

response
```

### 用上述網站說明該如何判斷 Header 中必須加上哪些資料才會正確


```python
headers = {'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.117 Safari/537.36'}
r = requests.get("https://www.zhihu.com/topic/19550228/hot", headers = headers)
response = r.text

response
```
