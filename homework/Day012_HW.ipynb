### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, 'html.parser')
```

```python
for d in soup.find(class_= 'part_list_2').find_all('h3'):
    print(d.find(class_='date').text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
catg = []
for d in soup.find(class_= 'part_list_2').find_all('h3'):
    if d.find(class_='tag').text not in catg:
        catg.append(d.find(class_='tag').text)
#print(catg)


all = []
for d in soup.find(class_= 'part_list_2').find_all('h3'):
    all.append(d.find(class_='tag').text)
    
for i in range(len(catg)):
    print(catg[i], all.count(catg[i]))
```

### ③ 哪一個類別的文章最多

```python
NUM_MAX = 0
CATG_MAX = ""

for i in range(len(catg)):
    if all.count(catg[i]) > NUM_MAX:
        NUM_MAX = all.count(catg[i])
        CATG_MAX = catg[i]
        
CATG_MAX
```
